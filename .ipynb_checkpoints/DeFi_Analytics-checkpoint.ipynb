```python
# Import necessary libraries
from web3 import Web3
from datetime import datetime
import pandas as pd
import json
from IPython.display import Image
```

```python
infura_url= 'https://mainnet.infura.io/v3/76210dc3bdfa4e94be8273cb29291eaa'
web3= Web3(Web3.HTTPProvider (infura_url))

# Check if the api is properly connected
web3.is_connected()
```

#### Exploring DAI (Ethereum's stablecoin)

```python
sc_address = "0x6B175474E89094C44Da98b954EedeAC495271d0F"
```

```python
with open ("./ABIs/erc20_abi.json") as f:
    abi = json.load (f)
```

```python
dai_contract = web3.eth.contract(address = sc_address, abi = abi)
```

```python
dai_contract.all_functions()
```

```python
print ("Name : ", dai_contract.functions.name().call())
```

```python
print ("Symbol_name: ", dai_contract.functions.symbol().call())
```

```python
print ("Decimals: ", dai_contract.functions.decimals().call())
```

```python
print ("Dai holdings: ", dai_contract.functions.balanceOf('0x60FaAe176336dAb62e284Fe19B885B095d29fB7F').call()*10**-18)
```

```python
print ("Max supply is: ", dai_contract.functions.totalSupply().call()*10**-18)
```
